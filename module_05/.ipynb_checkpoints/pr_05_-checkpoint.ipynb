Что было сделано:


**Предобработка:**
1. Прологорифмированы числовые переменные за исключением score_bki.
2. Для значение income выполнена min max нормализация относителльно рейтингов регионов.
3. На основании столбца app_date создано 2 новых признака бинарный - weekd_or_workd и категориальный - month.
4. Для категориальных признаков выполнен LabelEncoder.
5. Для бинарных признаков выполнен LabelEncoder.
6. Выполнена стандартизация данных StandartScaler, кроме income, так как его мы стандартизировали ранее.
7. Выполнена балансировка модели при помощи oversampling.


```python
# Загружаем библиотеки
from pandas import Series
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt
import seaborn as sns
import datetime

from datetime import datetime, timedelta
from sklearn.preprocessing import StandardScaler

from sklearn.feature_selection import f_classif, mutual_info_classif
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression


from sklearn.metrics import confusion_matrix
from sklearn.metrics import auc, roc_auc_score, roc_curve
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python

def show_confusion_matrix():
    tn, fp, fn, tp = confusion_matrix(y_valid, y_pred).ravel()
    names = ['TP','FP','FN','TN']
    counts = tp,fp,fn,tn
    labels = [f"{i}\n{j}" for i, j in zip(names,counts)]
    plt.figure()
    sns.heatmap(np.array(counts).reshape(2,2), annot=np.array(labels).reshape(2,2),
                annot_kws={"size": 15}, fmt='', cmap= 'BuPu', cbar = False,
                xticklabels = ['Дефолтный','Не дефолтный'], yticklabels= ['Дефолтный','Не дефолтный'])
    plt.title('Матрица ошибок')
    plt.show()

def show_roc_curve():
    probs = model.predict_proba(X_valid)
    probs = probs[:,1]
    fpr, tpr, threshold = roc_curve(y_valid, probs)
    roc_auc = roc_auc_score(y_valid, probs)
    plt.figure()
    plt.plot([0, 1], label='Baseline', linestyle='--')
    plt.plot(fpr, tpr, label = 'Regression')
    plt.title('Logistic Regression ROC AUC = %0.3f' % roc_auc)
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend(loc = 'lower right')
    plt.show()
    
def show_metrics():
    result = pd.Series({
            'accuracy' : accuracy_score(y_valid, y_pred),
            'precision' : precision_score(y_valid, y_pred),
            'recall' : recall_score(y_valid, y_pred),
            'F1-score' : f1_score(y_valid, y_pred)
        })
    return result
```

В целом с пропусками покончено, теперь разделим наши признаки на бинарные, категориальные и числовые для простоты работы и пробежимся по каждому, для оценки выбросов и уникальных значений, построении визуализации и осмотра на наличие ошибочных данных.


```python
#Для разбиения признаков по категориям посмотрим на уникальные значение признаков и их количество
data.agg({'nunique', lambda x: x.unique()}).transpose()
```

В данных есть признак app_date. Преобразуем его в читабельный вид. Создадим на его основе дополнительные признаки категориальный - столбец с месяцами и бинарный - столбец выходной или будний день.

```python
data['month'] = data['app_date'].apply(lambda x : datetime.strptime(x, '%d%b%Y').month)
data['weekd_or_workd'] = data['app_date'].apply(
    lambda x: 0 if (datetime.strptime(x, '%d%b%Y').weekday() < 5)
                and (datetime.strptime(x, '%d%b%Y').month==1 and datetime.strptime(x, '%d%b%Y').day > 8) 
                else 1)
```

```python
'''Разделим на категории'''
# Бинарные переменные
bin_cols = ['sex', 'car', 'car_type', 'good_work', 'foreign_passport', 'weekd_or_workd']
# Категориальные переменные
cat_cols = ['education', 'home_address', 'work_address', 'region_rating', 'sna', 'first_time', 'month']
# Числовые переменные
num_cols = [ 'income', 'age', 'bki_request_cnt', 'decline_app_cnt', 'score_bki']
```

# Числовые переменные

```python
for column in num_cols:
    plt.figure()
    sns.distplot(data[column], kde = False, rug=False)
    plt.title(column)
    plt.show()
```

По графикам видим, что признаки income, age, desline_app_cnt, bki_request_cnt имеют тяжелые правые хвосты. Эти признаки мы прологорифмируем. По графику признака score_bki можно сказать что данные распределены приблизительно нормально.

```python
for column in num_cols[:-1]:
    plt.figure()
    sns.distplot(np.log(data[column]+1), kde = False, rug=False)
    plt.title(column)
    plt.show()
```

В целом логарифмирование положительно повлияло на признаки, в особенности на income и age. Значения признака income сильно выделяются по сравнению с остальными, необходимо выполнить нормализацию данных. 

```python
# Произведем лог-преобразования с некоторыми числовыми признаками 
for column in num_cols[:-1]:
    data[column] = np.log(data[column] + 1)
```

```python
# посмотрим на распределение income по region_rating
fig, ax = plt.subplots(figsize = (14, 4))
sns.boxplot(x='region_rating', y='income', 
                data=data,
               ax=ax)
plt.xticks(rotation=45)
plt.show()
```

Видим что значения зарплат разнится по регионам. Выполним min max нормализацию относителльно регионов, чтобы не потерять эти данные.

```python
# Находим среднее значение по городам
min_income_log = data.groupby(by='region_rating')['income'].min()

# Находим стандартное отклонение значений Ranking по городам
max_income_log = data.groupby(by='region_rating')['income'].max()

# Добавляем колоки в наш датафрейм с этими значениями относительно каждого города
data['min_income'] = data['region_rating'].apply(lambda x: min_income_log[x])
data['max_income'] = data['region_rating'].apply(lambda x: max_income_log[x])

# Выполняем нормализацию Ranking в новый столбец
data['income'] = (data['income'] - data['min_income'])/(data['max_income']-data['min_income'])
```

```python
# Снова посмотрим на распределение income по region_rating
fig, ax = plt.subplots(figsize = (14, 4))
sns.boxplot(x='region_rating', y='income', 
                data=data,
               ax=ax)
plt.xticks(rotation=45)
plt.show()
```

Отлично, зависимость после нормализации сохранилась. А теперь построим боксплоты для численных переменных относительно признака default

# Подготовка к ML

```python
# Разделим данные на train и test
train_data = data[data['default'].isna() == False]
test_data = data[data['default'].isna() == True]
train_data
```

###  dummy-кодирование категориальных переменных

```python
X_cat_train = OneHotEncoder(sparse = False).fit_transform(train_data[cat_cols].values)
X_cat_test = OneHotEncoder(sparse = False).fit_transform(test_data[cat_cols].values)
```

### Нормирование числовых признаков

Выполняем стандартизацию данных с помощью StandartScaler, кроме столбца income, так как его мы стандартизировали ранее.

```python
# Стандартизация числовых переменных
scaler = StandardScaler()
X_num_train = scaler.fit_transform(train_data[num_cols[1:]].values)
X_num_test = scaler.transform(test_data[num_cols[1:]].values)
```

### Объединим все данные

```python
X_train = np.hstack([X_num_train, train_data[num_cols[0:1]].values, train_data[bin_cols].values, X_cat_train])
X_test = np.hstack([X_num_test, test_data[num_cols[0:1]].values, test_data[bin_cols].values, X_cat_test])
y_train = train_data['default'].values
```

```python
# Разделим данные для обучения следующим образом:
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.20, random_state=RANDOM_SEED)
train_data.shape, test_data.shape, X_train.shape, X_valid.shape
```

# Построение модели

```python
'''Логистическая регрессия'''
model = LogisticRegression(max_iter = 2000, random_state=RANDOM_SEED)
model.fit(X_train, y_train)
y_pred = model.predict(X_valid)
```

Посмотрим на Roc-кривую

```python
show_roc_curve()
```

Мы видим что значение ROC_AUC достаточно высоко, что означает о состоятельности модели, как классификатора данных. 

Для более точной оценки воспользуемся confusionm-matrix. На ней уже детально просмотрим наши полученные 'дефолтные' данные.

```python
show_confusion_matrix()
```

```python
show_metrics()
```

По матрице видим, что истинно положительных значений 8870 из 13022 реальных (около 68%), а истинно отрицательных 8642 из 13003 (около 66%). В целом по метрикам можно сказать, что модель неплохо ведет себя на сбалансированных данных. Попробуем улучшить результат при помощи регуляризации.


```python
model = LogisticRegression(random_state=RANDOM_SEED)
C = np.logspace(-1, 4, 15)
iter_ = 50
epsilon_stop = 1e-3
param_grid = [
    {'penalty': ['l1'], 
     'C': C,
     'solver': ['liblinear', 'lbfgs'], 
     'class_weight':['none', 'balanced'], 
     'multi_class': ['auto','ovr'], 
     'max_iter':[iter_],
     'tol':[epsilon_stop]},
    {'penalty': ['l2'], 
     'C': C,
     'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'], 
     'class_weight':['none', 'balanced'], 
     'multi_class': ['auto','ovr'], 
     'max_iter':[iter_],
     'tol':[epsilon_stop]},
    {'penalty': ['none'], 
     'C': C,
     'solver': ['newton-cg', 'lbfgs', 'sag', 'saga'], 
     'class_weight':['none', 'balanced'], 
     'multi_class': ['auto','ovr'], 
     'max_iter':[iter_],
     'tol':[epsilon_stop]}
    ]
gridsearch = GridSearchCV(model, param_grid, scoring='f1', n_jobs=-1)
gridsearch.fit(X_train, y_train)
model = gridsearch.best_estimator_
## печатаем параметры
best_parameters = model.get_params()
```

```python
best_parameters
```

```python
# Обучаем модель

model.fit(X_train, y_train)

y_pred_prob = model.predict_proba(X_valid)[:,1]
y_pred = model.predict(X_valid)
```

При анализе метрик можно заметить уменьшение precision и увеличение recall и f1-score, что тоже является положительным моментом. Также благодаря регуляризации мы избавились от возможного переобучения нашей модели.
