```python
# Функция которая удаляет выбросы
def waste_clean(data, column):
    perc25 = data[column].quantile(0.25)
    perc75 = data[column].quantile(0.75)
    IQR = perc75 - perc25
    maximum = perc75 + 1.5*IQR
    minimum = perc25 - 1.5*IQR
    return data[column].apply(lambda x: np.nan if x > maximum or x < minimum else x)
```

# Feature Engineering



### restalling_year

Это признак означающий количество лет после последнего рестайлинга. По сути он не может быть отрицательным.

```python
data['restallingYear'] = (data['productionDate'] - data['modelDate'])
data['restallingYear'].describe()
```

Видим, что здесь появиолсь отрицательное число, а также есть выброс в виде 46 лет. Посмотрим на эти данные ближе.

```python
data[data['restallingYear'] == -2]
```

Скорее всего здесь просто перепутаны даты местами. Возьмем разницу по модулю

```python
data['restallingYear'] = abs(data['restallingYear'])
```

```python
data['restallingYear'].value_counts()
```

По распределению видно, что большинство машин с более 'свежим' рестайлингом. Объединим значения с 10+ в 10.

```python
data['restallingYear'] = data['restallingYear'].apply(lambda x: 10 if x>10 else x)
```

# Tunned CatBoost + CrossValidation

```python
def cat_model(y_train, X_train, X_test, y_test):
    model = CatBoostRegressor(iterations = 6000,
                              learning_rate = 0.15,
                              eval_metric='MAPE',
                              random_seed = RANDOM_SEED,
                              one_hot_max_size=12,
                              random_strength=1)
    model.fit(X_train, y_train,
              cat_features=cat_features_ids,
              eval_set=(X_test, y_test),
              verbose=False,
              use_best_model=True,
              plot=False)
    
    return(model)


def mape(y_true, y_pred):
    return np.mean(np.abs((y_pred-y_true)/y_true))
```

```python
N_FOLDS = 5
submissions = pd.DataFrame(0,columns=["sub_1"], index=sample_submission.index) # куда пишем предикты по каждой модели
score_ls = []
splits = list(KFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_SEED).split(X, y))

for idx, (train_idx, test_idx) in tqdm(enumerate(splits), total=N_FOLDS):
    # use the indexes to extract the folds in the train and validation data
    train_idx = train_idx.reshape(64316)
    test_idx = test_idx.reshape(16079)
    X_train, y_train, X_valid, y_valid = X.iloc[train_idx], y.iloc[train_idx], X.iloc[test_idx], y.iloc[test_idx]
    # model for this fold
    model = cat_model(y_train, X_train, X_valid, y_valid)
    # score model on test
    test_predict = model.predict(X_valid)
    test_score = mape(y_valid, test_predict)
    score_ls.append(test_score)
    print(f"{idx+1} Fold Test MAPE: {mape(y_valid, test_predict):0.3f}")
    # submissions
    submissions[f'sub_{idx+1}'] = model.predict(X_sub)
    model.save_model(f'catboost_fold_{idx+1}.model')
    
print(f'Mean Score: {np.mean(score_ls):0.3f}')
print(f'Std Score: {np.std(score_ls):0.4f}')
print(f'Max Score: {np.max(score_ls):0.3f}')
print(f'Min Score: {np.min(score_ls):0.3f}')

```

Попробуем сравнить несколько комбинаций простых моделей, с подбором параметров.

```python
def compute_meta_feature_mean(clf, X_train, X_test, y_train, cv):
    """    Эта функция подсчитывает признаки для мета-классификатора.
    Они являются вероятностями классов при решении задачи многоклассовой классификации.   
    :arg clf: классификатор    
    :args X_train, y_train: обучающая выборка    
    :arg X_test: признаки тестовой выборки    
    :arg cv: класс, генерирующий фолды (KFold)    
    :returns X_meta_train, X_meta_test: новые признаки для обучающей и тестовой выборок    """
    n_classes = len(np.unique(y_train))
    X_meta_train = np.zeros((len(X_train), n_classes), dtype=np.float32)
    X_meta_test = np.zeros((len(X_test), n_classes), dtype=np.float32)
    for train_fold_index, predict_fold_index in cv.split(X_train):
        X_fold_train, X_fold_predict = X_train[train_fold_index], X_train[predict_fold_index]
        y_fold_train = y_train[train_fold_index]

        folded_clf = clone(clf)
        folded_clf.fit(X_fold_train, y_fold_train)

        X_meta_train[predict_fold_index] = folded_clf.predict_proba(X_fold_predict)
        
        X_meta_test += folded_clf.predict_proba(X_test)

    X_meta_test /= cv.n_splits

    return X_meta_train, X_meta_test

def generate_meta_features(classifiers, X_train, X_test, y_train, cv):
   
    features = [
        compute_meta_feature(clf, X_train, X_test, y_train, cv)
        for clf in tqdm(classifiers)
    ]
    
    stacked_features_train = np.hstack([
        features_train for features_train, features_test in features
    ])

    stacked_features_test = np.hstack([
        features_test for features_train, features_test in features
    ])
    
    return stacked_features_train, stacked_features_test
```

### получаем метапризнаки:

```python
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)


stacked_features_train, stacked_features_test = generate_meta_features([
    RandomForestClassifier(n_estimators=300, n_jobs=-1),
    ExtraTreesClassifier(n_estimators=300, n_jobs=-1)
], X_train, X_test, y_train, cv)
```
