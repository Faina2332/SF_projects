### enginePower

### engineDisplacement 
Посмотрим на пропуски в engineDisplacement

```python
data['engineDisplacement'] = data['engineDisplacement'].apply(lambda x: int(x.split(
    ' ')[0].replace('.', '')) if len(x.split(' ')[0]) > 0 else np.nan)
```

```python
data[(data['engineDisplacement'].isna()==True)]
```

Здесь мы видим много пропусков. Можно попробовать заполнить их в соответствии enginePower

```python
group_mean = data.groupby('enginePower')['engineDisplacement'].mean()
for i in data[(data['engineDisplacement'].isna()==True)].index:
    power = data['enginePower'][i]
    data['engineDisplacement'][i] = round(group_mean[power],0)
```

Посмотрим на оставшиеся пропуски

```python
data[(data['engineDisplacement'].isna()==True)]
```

Заполним минимальным значением, так как enginePower здесь минимально возможный, так как машина электрическая.

```python
data['engineDisplacement'][(data['engineDisplacement'].isna()==True)] = int(group_mean.min())
```

```python
data['engineDisplacement'] = data['engineDisplacement'].apply(lambda x: int(x))
```

# Feature Engineering



### restalling_year

Это признак означающий количество лет после последнего рестайлинга. По сути он не может быть отрицательным.

```python
data['restallingYear'] = (data['productionDate'] - data['modelDate'])
data['restallingYear'].describe()
```

Видим, что здесь появиолсь отрицательное число, а также есть выброс в виде 46 лет. Посмотрим на эти данные ближе.

```python
data[data['restallingYear'] == -2]
```

Скорее всего здесь просто перепутаны даты местами. Возьмем разницу по модулю

```python
data['restallingYear'] = abs(data['restallingYear'])
```

```python
data['restallingYear'].value_counts()
```

По распределению видно, что большинство машин с более 'свежим' рестайлингом. Объединим значения с 10+ в 10.

```python
data['restallingYear'] = data['restallingYear'].apply(lambda x: 10 if x>10 else x)
```

### carAge
Добавим признак возраст автомобиля с округлением до 5

```python
data['carAge'] = 2021 - data['productionDate']
data['carAge'] = round(data['carAge']//5)*5
data['carAge'].value_counts()
```

### kmАge

Добавим переменную kmАge (кол-во км за год эксплуатации) 

```python
data['kmАge'] = data['mileage'][(data['mileage']>0.0)&(data['productionDate']<2021.0)]/(2021 - data['productionDate'][(data['mileage']>0.0)&(data['productionDate']<2021.0)])
```

Там где получились пустые значения пробег был нулевым. Следовательно, обозначим новый признак за ноль.

```python
data['kmАge'][data['kmАge'].isna()==True] = 0
```

```python
data['kmАge'].nunique()
```

```python
data['kmАge'].value_counts()
```

Получилось большое количество разных данных, округлим значения кратно тысячам, как сделано в mileage.

```python
data['kmАge'] = data['kmАge'].apply(lambda x: int(round(x/100000,2)*100000))
```

```python
data['kmАge'].nunique()
```

Добавим новые призрнаки в наш список числовых признаков

# Tunned CatBoost + CrossValidation

```python
def cat_model(y_train, X_train, X_test, y_test):
    model = CatBoostRegressor(iterations = 5000,
                              learning_rate = 0.15,
                              eval_metric='MAPE',
                              random_seed = RANDOM_SEED,
                              one_hot_max_size=12,
                              random_strength=1)
    model.fit(X_train, y_train,
              cat_features=cat_features_ids,
              eval_set=(X_test, y_test),
              verbose=False,
              use_best_model=True,
              plot=False)
    
    return(model)


def mape(y_true, y_pred):
    return np.mean(np.abs((y_pred-y_true)/y_true))
```

```python
N_FOLDS = 5
submissions = pd.DataFrame(0,columns=["sub_1"], index=sample_submission.index) # куда пишем предикты по каждой модели
score_ls = []
splits = list(KFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_SEED).split(X, y))

for idx, (train_idx, test_idx) in tqdm(enumerate(splits), total=N_FOLDS):
    # use the indexes to extract the folds in the train and validation data
    train_idx = train_idx.reshape(64316)
    test_idx = test_idx.reshape(16079)
    X_train, y_train, X_valid, y_valid = X.iloc[train_idx], y.iloc[train_idx], X.iloc[test_idx], y.iloc[test_idx]
    # model for this fold
    model = cat_model(y_train, X_train, X_valid, y_valid)
    # score model on test
    test_predict = model.predict(X_valid)
    test_score = mape(y_valid, test_predict)
    score_ls.append(test_score)
    print(f"{idx+1} Fold Test MAPE: {mape(y_valid, test_predict):0.3f}")
    # submissions
    submissions[f'sub_{idx+1}'] = model.predict(X_sub)
    model.save_model(f'catboost_fold_{idx+1}.model')
    
print(f'Mean Score: {np.mean(score_ls):0.3f}')
print(f'Std Score: {np.std(score_ls):0.4f}')
print(f'Max Score: {np.max(score_ls):0.3f}')
print(f'Min Score: {np.min(score_ls):0.3f}')

```
